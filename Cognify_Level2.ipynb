```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/Dataset .csv')
```

```python
df.info()
```

```python
df.head(5)
```

```python
table_booking_count=df['Has Table booking'].value_counts()
online_delivery_count = df['Has Online delivery'].value_counts()

total_restaurants=len(df)

percentage_table_booking = (table_booking_count.get('Yes', 0) / total_restaurants) * 100
percentage_online_delivery = (online_delivery_count.get('Yes', 0) / total_restaurants) * 100

print("Percentage of Restaurants Offering Table Booking:", percentage_table_booking)
print("Percentage of Restaurants Offering Online Delivery:", percentage_online_delivery)
```

```python
average_rating_with_booking = df[df['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_booking = df[df['Has Table booking'] == 'No']['Aggregate rating'].mean()

print("Average Rating for Restaurants with Table Booking:", average_rating_with_booking)
print("Average Rating for Restaurants without Table Booking:", average_rating_without_booking)
```

```python
online_delivery_percentage_by_price_range = df.groupby('Price range')['Has Online delivery'].apply(lambda x: (x == 'Yes').mean() * 100)

print("Online Delivery Percentage by Price Range:")
print(online_delivery_percentage_by_price_range)
```

```python
price_range_counts = df['Price range'].value_counts()


most_common_price_range = price_range_counts.idxmax()


print("Most Common Price Range:", most_common_price_range)
```

```python
average_rating_by_price_range = df.groupby('Price range')['Aggregate rating'].mean()

print("Average Rating for Each Price Range:")
print(average_rating_by_price_range)
```

```python
rating_color_mapping = {
    'Dark Green': 5,
    'Green': 4,
    'Yellow': 3,
    'Orange': 2,
    'White': 1,
    'Red': 0
}

price_range_rating_mapping = {
    3: 'Dark Green',
    4: 'Green',
    2: 'Yellow',
    1: 'Orange'
}


highest_avg_rating_price_range = average_rating_by_price_range.idxmax()
print(highest_avg_rating_price_range)

color_for_highest_avg_rating = price_range_rating_mapping[highest_avg_rating_price_range]

print("Color representing the highest average rating among different price ranges:", color_for_highest_avg_rating)
```
